```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
USA = Test.merge(Vax, on=['state'])
```

```python
Death=pd.read_csv("/content/DeathUSA.csv")
```

Example: Full Moderna Set v. Hospitalization

```python
md = np.mean(USA.ModernaFull) - np.mean(USA.hospitalized)
md
```

```python
se = np.sqrt(np.var(USA.ModernaFull)/len(USA.ModernaFull) + np.var(USA.hospitalized)/len(USA.hospitalized))
```

```python
lcl,ucl = md-2*se,md+2*se
print(lcl, ucl)
```

```python
ModernaHospital = USA[["ModernaFull", "hospitalized"]]
```

```python
sns.pairplot(ModernaHospital)
```

At least one dose

```python
plt.scatter(USA['Moderna'], USA['hospitalized'])
```
